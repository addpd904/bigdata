```python
# numpy(numerical python);for calculating matrix and so on;base on c
# ndarray (n,dimension,array)
import numpy as np
score=np.array([[1,3],
               [5,7],
               [9,2]],dtype=np.int64)
# 一、basic grammer
# a.property
# row and column(2,3) 
score.shape
# dimension
score.ndim
# type 
# int default :int32
score.dtype
# size of every item
score.itemsize
# b.method
# 1.create 0 and 1 array
np.zeros((3,3))
np.ones((2,4))
# 2.array()
# asarray():shallow copy
# copy():deep copy
scores=[[1,3],[5,7],[9,2]]
nparray=np.copy(scores)

# 3.generate fixed range array
# linespace:equidistance;5:numbers
nparray2=np.linspace(0,10,5)
# arrange();5 is step 
nparray3=np.arange(0,10,5)
# 4.set distribution /ˌdɪstrɪˈbjuːʃn/
# *uniform distribution
nparray3=np.random.uniform(low=-1,high=1,size=100000)
# data visualization
import matplotlib.pyplot as plt
plt.figure()
plt.hist(nparray3,1000)
plt.show()
# *normal distribution
# basic conception:standard deviation and average value
# loc is average value,scale is standard deviation
nparray4=np.random.normal(loc=2,scale=0.1,size=1000000)
# data visualization
plt.figure()
plt.hist(nparray4,1000)
plt.show()
```
