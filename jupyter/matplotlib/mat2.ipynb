```python
# 一、scatter
import matplotlib.pyplot as plt

import random
# 1.ready data
x=[random.randrange(0,200) for i in range(100)]
y=[random.randrange(0,500) for i in range(100)]
# 2.create figure
plt.figure(figsize=(20,20),dpi=60)
plt.scatter(x,y)
plt.show()

```

```python
# 二、bar
# 1.ready data
name=['zs','zjl','cyx','dzq']
en_scores=[88,66,99,100]
ma_scores=[48,86,49,80]

# 2.create figure
plt.figure(figsize=(20,8),dpi=60)
# 3.draw bar and add a label for legend
# [1,4,7,10] is x-axis value
plt.bar([1,4,7,10],en_scores,color=['pink','pink','pink','pink'],label='en') 
plt.bar([2,5,8,11],ma_scores,color=['red','red','red','red'],label='ma') 
# 4.show the legend
plt.legend()
# 5.add the ticks and label for x;zs ls
# [1.5,4.5,7.5,10.5] is ticks and corresponding to label
plt.xticks([1.5,4.5,7.5,10.5],name)
# 6.title
plt.title('scores')
# 7.
plt.grid(linestyle='--',alpha=0.5)
plt.show()
```

```python
# 三、histogram
# basic:conception:class number and class interval
# 1.ready data and genarate figure
data=[random.randrange(1,60) for i in range(50)]
plt.figure(figsize=(20,8),dpi=50)
# 2.calculate the basic patameter
# clInte: class interval
clInte=10
# clInte: class number
clNum=int((max(data)-min(data))/clInte)+1
# 3.draw histogram
# bins: class number;density:frequency
plt.hist(data,bins=clNum,density=True)
# 4.modify ticks for adpating histogram
plt.xticks(range(0,61,clInte))
plt.show()
```

```python
# 四、pie
# 1.ready data and create figure
label=['male','female']
count=[400,180]
# 2.draw
plt.pie(count,labels=label,colors=['blue','pink'],autopct='%1.2f%%')
# show legend
plt.legend()
plt.show()

```
