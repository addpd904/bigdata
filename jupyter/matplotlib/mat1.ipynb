```python
import matplotlib.pyplot as plt
%matplotlib inline

# 一、basic grammer
plt.figure   #create a figure
# x axis and y axis
plt.plot([1,2,4,5],[5,9,4,6])
plt.show()
#  二、Three-tier structure
# a.container layer:1.canvas(drawing board) 2.figure 3.axes
# 3.plt.subplots()

# b.Auxiliary /ɔːɡˈzɪliəri/ display layer(base on container layer)
# tick,mesh generation,title,legend

# c.image layer
# set the style and color,bar, histogram,plot,scatter,pie


```

```python
# 三、plot
# a.basic
# dpi:dot per inch(articulation)
plt.figure(figsize=(20,8),dpi=80)
plt.plot([1,2,3,4],[24,20,25,37],color='pink',label='shanghai')
plt.plot([1,2,4,5],[14,10,35,17],linestyle='--',label='beijing')
# show legend
plt.legend()
# save the image
plt.savefig(r'E:\programme\Python\practice\others\plot.png')
# caution:show()will release all resource
plt.show()
```

```python
# b.Auxiliary /ɔːɡˈzɪliəri/ display layer
import random
# 1.ready data
# return a list
x=range(60)
# return a list,uniform 
y_shanghai=[random.uniform(15,18) for i in x]
# 2. draw plot
plt.figure(figsize=(20,8),dpi=70)

plt.plot(x,y_shanghai)

# 3.DIY tick
x_label=[f'11 dian{i}fen' for i in range(0,61,5)]
plt.xticks(range(0,61,5),x_label)
plt.yticks(range(0,36,5))
# change the font for supporting chinese
# 4.add grid
plt.grid(linestyle='--',alpha=0.5)
# 5.add description for x,y
plt.xlabel('time')
plt.ylabel('temperature')
plt.title('shanghai temperature change')

plt.show()

```

```python
# c.subplots,OOP(object-oriented programming)
#  return figure and axes
# 1.ready data
# return a list
x=range(60)
# return a list,uniform 
y_shanghai=[random.uniform(15,18) for i in x]
y_beijing=[random.uniform(1,4) for i in x]
# 2. draw plot
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(20,8),dpi=70)
axes[0].plot(x,y_shanghai)
axes[1].plot(x,y_beijing)

# 3.DIY tick
x_label=[f'{i}fen' for i in range(0,61,10)]
axes[0].set_xticks(range(0,61,5),x_label)
axes[0].set_yticks(range(0,36,5))
axes[1].set_xticks(range(0,61,5),x_label)
axes[1].set_yticks(range(0,36,5))

# 4.add grid
axes[0].grid(linestyle='-',alpha=0.5)
axes[1].grid(linestyle='--',alpha=0.5)
# 5.add description for x,y
axes[0].set_xlabel('time')
axes[0].set_ylabel('temperature')
axes[0].set_title('shanghai temperature change')

axes[1].set_xlabel('time')
axes[1].set_ylabel('temperature')
axes[1].set_title('beijing temperature change')

plt.show()
```
